```python
# It seems there was an issue with the previous approach. Let's debug and fix it.

import random
import pandas as pd

file_path_csv = './experiment_conditions.csv'
df_loaded_csv = pd.read_csv(file_path_csv)
display(df_loaded_csv)

SEARCHED = False
while not SEARCHED:
    random_number = random.randint(0, 26)
    random_col = random.choice(df_loaded_csv.columns)
    if pd.isna(df_loaded_csv.iloc[random_number][random_col]):
        df_loaded_csv.loc[random_number, random_col] = 'Marked'
        print('Marked',random_col, random_number)
        SEARCHED = True
    else:
        print('Already marked',random_col, random_number)
df_loaded_csv.to_csv(file_path_csv, index=False)

```
